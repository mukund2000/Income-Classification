## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.arange(1,10,0.5)
plt.plot(x,np.sin(x),color="black",linestyle="dashdot")
plt.xlabel("X",color="blue",size=30)
plt.ylabel("Y",color="purple",size=30)
plt.title("Sinosodial Graph",color="black",size=35)
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
Cities=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
Clothing=np.array([82,78,48,46,39])
Equipment=np.array([48,31,51,29,18])
Accessories=np.array([15,20,14,13,9])
plt.bar(Cities,Clothing,color="red",width=0.8,bottom=Equipment+Accessories,label="clothing")
plt.bar(Cities,Equipment,color="blue",width=0.8,bottom=Accessories,label="equipment")
plt.bar(Cities,Accessories,color="grey",width=0.8,label="accesories")
plt.xlabel("Cities",color="purple",size=30)
plt.ylabel("Neccessities",color="purple",size=30)
plt.title("Cities Vs Neccessities",size=30)
plt.legend(loc="upper right")
plt.xticks(rotation=90)
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
print(sns.get_dataset_names(),sep=" ")
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
survival=sns.countplot(x='pclass',hue='survived',palette='twilight',data=df)
survival.set(title='Survival of passengers with respect to passenger class',xlabel='Pclass',ylabel='Survived')
plt.legend(loc='upper right')
plt.show()
```

## Q.10 For Tips Dataset
- Load the dataset using seaborn
- Draw the Strip Plot between 'total bill' and 'day'
- Rotate xticks with an angle of 45 Degree

```python
np.random.seed(50)
sns.stripplot(x='day',y='total_bill',data=df_tips,palette='Set1',alpha=0.8)
plt.title("Day vs Total Bill",size=30)
plt.xlabel('Day',size=30)
plt.ylabel('Total Bill',size=30)
plt.xticks(rotation=45)
plt.show()
```
